```r
library(dplyr)
library(MASS)
library(olsrr)
library(car)
library(wiqid) # standardizing variables

rm(list = ls())

setwd("C:\\Users\\Dick Sang\\Desktop\\5. Data Analytics\\3. PolyU RA\\1. Projects\\3. Cust Value Chain Analysis\\4. Regression\\7. 2020-02-20\\2. analysis")
data <- read.csv("3. Data Input for Analysis.csv", stringsAsFactors=FALSE)

str(data)

data2 <- data %>%
         mutate(WC_Normalize_Optimistic = standardize(WC_Normalize_Optimistic, center = TRUE, scale = TRUE)) %>% 
         mutate(WC_Normalize_Affective = standardize(WC_Normalize_Affective, center = TRUE, scale = TRUE)) %>%
         mutate(Opt_stdize_n_Functional = WC_Normalize_Optimistic * WC_Normalize_Functional) %>%
         mutate(Opt_stdize_n_Experimental = WC_Normalize_Optimistic * WC_Normalize_Experimental) %>%
         mutate(Opt_stdize_n_Symbolic = WC_Normalize_Optimistic * WC_Normalize_Symbolic) %>%
         mutate(Opt_stdize_n_Cost = WC_Normalize_Optimistic * WC_Normalize_Cost) %>%
         mutate(Aff_stdize_n_Functional = WC_Normalize_Affective * WC_Normalize_Functional) %>%
         mutate(Aff_stdize_n_Experimental = WC_Normalize_Affective * WC_Normalize_Experimental) %>%
         mutate(Aff_stdize_n_Symbolic = WC_Normalize_Affective * WC_Normalize_Symbolic) %>%
         mutate(Aff_stdize_n_Cost = WC_Normalize_Affective * WC_Normalize_Cost)
```

```r
model <- glm(Adj_Sales ~
               WC_Normalize_Functional +
               WC_Normalize_Experimental +
               WC_Normalize_Symbolic+
               WC_Normalize_Cost
             , data = data2)

summary(model)

null_model <- glm(Adj_Sales ~ 1, data = data2)

# model1 <- stepAIC(null_model, direction="forward",scope=list(upper=model,lower=null_model))
model1 <- stepAIC(model, direction="backward")

summary(model1)
```

```r
##########################################################
# fine-tune final model
##########################################################
model2 <- lm(Adj_Sales ~ WC_Normalize_Functional +
                         WC_Normalize_Experimental +
                         #WC_Normalize_Symbolic+
                         #WC_Normalize_Cost+
                         WC_Normalize_Optimistic +
                         WC_Normalize_Affective
                            , data = data2
                    )

# plot(data2$WC_Normalize_Functional, data2$Adj_Sales)

summary(model2)
```
